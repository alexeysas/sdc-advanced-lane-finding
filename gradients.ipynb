```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import pickle
```

```python
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    grad = cv2.Sobel(gray, cv2.CV_64F, 1 if orient == 'x' else 0, 1 if orient == 'y' else 0, ksize=sobel_kernel)
    grad = np.abs(grad)
    max_grad = np.max(grad)
    grad = np.uint8(grad * 255 / max_grad)
    mask = np.zeros_like(grad)
    mask[(grad > thresh[0]) & (grad < thresh[1])] = 1
    return mask

def mag_thresh(image, sobel_kernel=3, mag_thresh=(0, 255)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    grad_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    grad_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    mag = np.sqrt(grad_x * grad_x + grad_y * grad_y)
    max_mag = np.max(mag)
    mag = np.uint8(mag * 255 / max_mag)
    mask = np.zeros_like(mag)
    mask[(mag > mag_thresh[0]) & (mag < mag_thresh[1])] = 1
    
    return mask

def dir_threshold(image, sobel_kernel=3, thresh=(0, np.pi/2)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    grad_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=sobel_kernel)
    grad_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=sobel_kernel)
    
    grad_x = np.abs(grad_x)
    grad_y = np.abs(grad_y)
        
    dir = np.arctan2(grad_y, grad_x)
    
    mask = np.zeros_like(dir)
    mask[(dir > thresh[0]) & (dir < thresh[1])] = 1
    
    return mask
```

```python
## Load image
img = mpimg.imread('test.png')
plt.imshow(img)
plt.show()
```

```python
grad1 = abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(20, 255))
plt.imshow(grad1, cmap='gray')
plt.show()
```

```python
grad1 = abs_sobel_thresh(img, orient='y', sobel_kernel=3, thresh=(20, 255))
plt.imshow(grad1, cmap='gray')
plt.show()
```

```python
grad2 = mag_thresh(img, sobel_kernel=31, mag_thresh=(50, 255))
plt.imshow(grad2, cmap='gray')
plt.show()
```

```python
grad3 = dir_threshold(img, sobel_kernel=15, thresh=(0.7, 1.3))
plt.imshow(grad3, cmap='gray')
plt.show()
```

```python
combined = np.zeros_like(grad3)
combined[((grad2 == 1) & (grad3 == 1))] = 1
plt.imshow(combined, cmap='gray')
plt.show()
```
